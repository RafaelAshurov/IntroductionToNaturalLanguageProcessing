# Imports

```python
import nltk
nltk.download('punkt')
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import spacy
```

# Data

```python
texts={
        1:"""When Sebastian Thrun started working on self-driving cars at 
        Google's headquarters in 2007, few people outside of the company took him 
        seriously. “I can tell you very senior CEOs of major American 
        car companies would shake my hand and turn away because I wasn’t 
        worth talking to,” said Thrun, in an interview with Recode earlier 
        this week.""",
        2:"""Harry Potter is a series of seven fantasy novels written by British author J. K. Rowling. 
            The novels chronicle the lives of a young wizard, Harry Potter, and his friends Hermione Granger and Ron Weasley, all of whom are students at Hogwarts School of Witchcraft and Wizardry.""",
        3:"""At a diner, Jules tells Vincent that he plans to retire from his life of crime, convinced that their 'miraculous' survival at the apartment was a sign of divine intervention. 
            While Vincent is in the bathroom, a couple, 'Pumpkin' and 'Honey Bunny', hold up the restaurant and demand Marsellus's briefcase."""
        }
```

# 1. nltk

### We will practice some of the options provided by the nltk library. Each of the following tasks must be solved using the nltk library.

## 1.a Split each of the texts into its <u>tokens</u>

```python
tokens = {}
for key in texts:
  tokens[key] = word_tokenize(texts[key])
tokens
```

## 1.b Split each of the texts into its <u>sentences</u>

```python
sentences = {}
for key in texts:
  sentences[key] = sent_tokenize(texts[key])
sentences
```

## 1.c For each text - split it into its <u>sentences</u> and then split each sentence into its <u>tokens</u>.

```python
sentences_tokens = {1:[], 2:[], 3:[]}
for key in texts:
    text_sentences = sent_tokenize(texts[key])
    for sentence in text_sentences:
        sentences_tokens[key] = sentences_tokens[key] + word_tokenize(sentence)
sentences_tokens
```

## 1.d For each text - split it into its <u>sentences</u> and then split each sentence into its <u>lemmas</u>.

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
sentences_lemmas = {}
wordnet_lemmatizer = WordNetLemmatizer()
for key in sentences_tokens:
  sentences_lemmas[key] = []
  for token in sentences_tokens[key]:
    sentences_lemmas[key].append(wordnet_lemmatizer.lemmatize(token))
sentences_lemmas
```

# 2. SpaCy

### Now we will practice using the SpaCy library. Each of the following tasks must be solved using the SpaCy library.

## 2.a Begin by loading SpaCy and tokenizing each of the texts

```python
nlp = spacy.load("en_core_web_sm")
tokens = {}
for key in texts:
  tokens[key] = [token.text for token in nlp(texts[key])]
tokens
```

## 2.b For each text - identify the all the verbs and adjectives. Hint: the Part of Speech (pos) of adjectives in SpaCy is "ADJ"

```python
verb_adj_seperation = {}
for key in texts:
    verb_adj_seperation[key] = {'verbs': [], 'adjectives': []}
    verb_adj_seperation[key]['verbs'] = [token.text for token in nlp(texts[key]) if token.pos_ == "VERB"]
    verb_adj_seperation[key]['adjectives'] = [token.text for token in nlp(texts[key]) if token.pos_ == "ADJ"]
verb_adj_seperation
```

## 2.c Separate each of the texts into its sentences.

```python
sentences = {}
for key in texts:
  sentences[key] = [sent.text for sent in nlp(texts[key]).sents]
sentences
```

## 2.d Separate each of the texts into its sentences, and then identify all the lemmas in each sentence

```python
sentences_lemmas = {}
for key in texts:
    sentences_lemmas[key] = []
    for sent in [sent.text for sent in nlp(texts[key]).sents]:
        sentences_lemmas[key] = sentences_lemmas[key] + [token.lemma_ for token in nlp(sent)]
sentences_lemmas
```

## 2.e Separate each of the texts into its sentences, and then identify all the verbs and adjectives in each sentence

```python
verb_adj_seperation = {}
for key in texts:
    verb_adj_seperation[key] = []
    for sentence in [sent.text for sent in nlp(texts[key]).sents]:
      verb_adj_seperation[key].append({
          'original_sentence': sentence,
          'verbs': [token.text for token in nlp(sentence) if token.pos_ == "VERB"],
          'adjectives': [token.text for token in nlp(sentence) if token.pos_ == "ADJ"]
      })
verb_adj_seperation
```
